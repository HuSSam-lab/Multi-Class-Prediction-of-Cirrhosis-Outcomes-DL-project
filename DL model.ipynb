```python
train_data, validation_data = train_test_split(data,test_size=0.2)
train_data.describe(include='all')

CAT_COL = ['Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']
NUM_COL = ['N_Days', 'Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin', 'Stage']

train_data_encoded = pd.get_dummies(train_data[CAT_COL])
validation_data_encoded = pd.get_dummies(validation_data[CAT_COL])

train_data_CAT_NUM = pd.concat([train_data[NUM_COL], train_data_encoded, train_data['Status']],axis=1 )
valid_data_CAT_NUM = pd.concat([validation_data[NUM_COL], validation_data_encoded, validation_data['Status']],axis=1 )



target = train_data_CAT_NUM['Status'].map({'C':0, 'CL':1, 'D':2})
valid_target = valid_data_CAT_NUM['Status'].map({'C':0, 'CL':1, 'D':2})



scaler = StandardScaler()
# scaled_num_train_data = scaler.fit_transform(train_data[TRAIN_LIST])
train_data_CAT_NUM[NUM_COL] = pd.DataFrame(scaler.fit_transform(train_data_CAT_NUM[NUM_COL]))
valid_data_CAT_NUM[NUM_COL] = pd.DataFrame(scaler.fit_transform(valid_data_CAT_NUM[NUM_COL]))


inputs = train_data_CAT_NUM.replace({np.NaN:0,True:1,False:0})
valid_inputs = valid_data_CAT_NUM.replace({np.NaN:0,True:1,False:0})

target = train_data_CAT_NUM['Status'].replace({'C':0, 'CL':1, 'D':2})
valid_target = valid_data_CAT_NUM['Status'].replace({'D':0, 'C':1, 'CL':2})

# inputs.drop('Status', axis=1)
# valid_inputs.drop('Status', axis=1)
inputs.drop(columns='Status',inplace=True)
valid_inputs.drop(columns='Status',inplace=True)
inputs
```

```python
model = tf.keras.Sequential([

            tf.keras.layers.Dense(10,activation='relu'),
            tf.keras.layers.Dense(10,activation='relu'),

            tf.keras.layers.Dense(3,activation='softmax')
    
])
model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=inputs,y=target,validation_data=(valid_inputs,valid_target) ,epochs=100,verbose=1,batch_size=50)
```

```python
import pickle 
with open('model', 'wb') as file: ### file name = model, wb = write byte, 
    pickle.dump(model, file)
```
